```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import coremltools as ct

# 1. CREATE DUMMY TRAINING DATA
data = {
    'hour': [6, 8, 10, 12, 14, 16, 18, 20, 22],
    'percent_complete': [0.0, 0.1, 0.25, 0.45, 0.60, 0.75, 0.90, 0.95, 1.0]
}
df = pd.DataFrame(data)

X = df[['hour']]
y = df['percent_complete']

# 2. TRAIN THE MODEL
# Ensure you are using scikit-learn <= 1.5.1
model = LinearRegression()
model.fit(X, y)

print(f"Model trained! Coefficient: {model.coef_}")

# 3. CONVERT TO COREML
# Fixed argument name: 'output_feature_names'
coreml_model = ct.converters.sklearn.convert(
    model, 
    input_features=['hour'], 
    output_feature_names='percent_complete'
)

# 4. SAVE THE FILE
coreml_model.save("StepPredictor.mlmodel")
print("Saved StepPredictor.mlmodel - Drag this into Xcode!")
```
